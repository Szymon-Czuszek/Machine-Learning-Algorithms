## Wprowadzenie do drzew decyzyjnych
Drzewo decyzyjne to narzędzie wspomagające podejmowanie decyzji, które wykorzystuje drzewiasty model decyzji i ich możliwych konsekwencji, w tym wyników zdarzeń losowych, kosztów zasobów i użyteczności. Jest to jeden ze sposobów wyświetlania algorytmu, który zawiera tylko warunkowe instrukcje sterujące.

###### Zalety
1. Drzewa decyzyjne są bardzo łatwe do wytłumaczenia ludziom
2. Drzewa decyzyjne lepiej odzwierciedlają ludzki proces decyzyjny niż inne metody regresji i klasyfikacji
3. Drzewa XGTrees mogą być wyświetlane graficznie i są łatwe do interpretacji nawet przez osoby niebędące ekspertami.
4. Drzewa mogą z łatwością obsługiwać predyktory jakościowe bez konieczności tworzenia zmiennych zastępczych.
###### Wady
1. Drzewa generalnie nie mają takiego samego poziomu dokładności predykcyjnej, jak niektóre inne metody regresji i klasyfikacji.


#### Jak działa drzewo?
1. Dzielimy przestrzeń predyktorów - czyli zbiór możliwych wartości dla: X1,X2, ...,Xp - na J odrębnych i nie nakładających się regionów: R1,R2, . . . , RJ .
2. Dla każdej obserwacji, która należy do regionu Rj, dokonujemy tej samej prognozy, która jest po prostu średnią wartości odpowiedzi dla obserwacji szkoleniowych w Rj .
3. Celem jest zminimalizowanie RSS
4. Stosuje odgórne, zachłanne podejście znane jako rekurencyjny podział binarny.
    - Odgórne, ponieważ zaczyna się na szczycie drzewa, a następnie sukcesywnie dzieli przestrzeń predyktorów
    - Każdy podział jest wskazywany przez dwie nowe gałęzie w dalszej części drzewa.
    - Jest zachłanne, ponieważ na każdym etapie procesu tworzenia drzewa dokonywany jest najlepszy podział na tym konkretnym etapie, zamiast patrzeć w przyszłość i wybierać podział, który doprowadzi do lepszego drzewa w jakimś przyszłym kroku.
5. Kontrolowanie wzrostu drzewa:
    - Uwzględnia wszystkie predyktory i wszystkie możliwe wartości punktu odcięcia
    - Oblicza RSS dla każdej możliwości
    - Wybiera tę z najmniejszym RSS
    - Kontynuuje do osiągnięcia kryteriów zatrzymania
6. Kroki przycinania
    - Wyhodowanie bardzo dużego drzewa
    - Przyciąć je, aby uzyskać optymalne drzewo
7. Kryteria zatrzymania
    - Minimalna liczba obserwacji w węźle wewnętrznym (Minimalna liczba obserwacji wymagana do dalszego podziału)
    - Minimalna liczba obserwacji w węźle liścia (Minimalna liczba obserwacji wymagana w każdym węźle po podziale)
    - Maksymalna głębokość (Maksymalne możliwe warstwy drzewa)

#### Typy drzew decyzyjnych
1. **Drzewo regresji**: Dla ciągłej ilościowej zmiennej docelowej. (Oparte na średniej)
Np. przewidywanie opadów deszczu, przewidywanie przychodów, itp.
2. **Drzewo klasyfikacyjne**: Dla dyskretnych kategorycznych zmiennych docelowych (Oparte na modzie)
Np. przewidywanie wysokiego lub niskiego poziomu, wygranej lub przegranej, zdrowego lub niezdrowego itp.

- Drzewa klasyfikacyjne są narzędziem używanym do określania przynależności różnych przypadków lub rzeczy do określonych kategorii na podstawie informacji zebranych na temat tych przypadków. Wykorzystuje się je w analizie danych, szczególnie w dziedzinie data mining, aby pomóc w zrozumieniu wzorców i relacji w danych. Te drzewa pomagają w przewidywaniu, które przypadki należą do których kategorii, bazując na różnych właściwościach lub zmiennych objaśniających, które są mierzone. To jedna z podstawowych metod używanych w zgłębianiu danych, co pomaga w odkrywaniu cennych informacji i trendów w danych.
- Drzewa klasyfikacyjne są rzadko używane w teorii prawdopodobieństwa i statystycznym rozpoznawaniu obrazów, ale znajdują szerokie zastosowanie w praktyce, np. w medycynie, naukach komputerowych, botanice i psychologii. Ich zaletą jest łatwość wizualizacji, co ułatwia interpretację w porównaniu z wynikami numerycznymi.


#### Różnice i podobieństwa
- Zarówno regresja, jak i klasyfikacja wykorzystują rekurencyjny podział binarny, wspomniany we wcześniejszej części pracy.
- W regresji RSS jest używany do decydowania o metodach podziału
- W klasyfikacji możemy użyć (Wskaźnik Giniego i entropia krzyżowa oznaczają czystość węzła):

    - Wskaźnik błędu klasyfikacji (Fragment danych, który zawiera wiele różnych klas lub kategorii, nazywamy obszarem zanieczyszczonym. Im większa różnorodność klas w tym fragmencie, tym większe jest zanieczyszczenie danych.)
    
$$
\text{Błąd klasyfikacji} = 1 - \max(P(y=k))
$$

    - Wskaźnik Giniego (Współczynnik Giniego to metryka używana do pomiaru nierówności w podziale danych, co pomaga w wyborze najlepszego kryterium podziału dla określonego węzła decyzyjnego. Wartość współczynnika Giniego mieści się w przedziale od 0 do 1, gdzie 0 oznacza, że wszystkie obiekty należą do tej samej klasy lub grupy, natomiast 1 oznacza, że podział jest maksymalnie nierównomierny, czyli każda klasa ma oddzielny obszar.)
    
$$
\text{Gini}(p) = 1 - \sum_{k=1}^{n} p_k^2
$$

    - Entropia krzyżowa i zysk informacyjny (W entropii, zysku informacyjnym i współczynniku Giniego chodzi o to samo - pomiar nierównomierności podziału danych. W tym podejściu, najpierw obliczamy entropię dla atrybutu decyzyjnego, czyli jak bardzo dane są rozproszone. Następnie obliczamy entropię podziału zbioru według tego atrybutu, czyli jak bardzo dane są rozproszone w każdym z podziałów. Zysk informacyjny to różnica między entropią atrybutu decyzyjnego a entropią podziału, co pokazuje, jak dużo informacji zyskujemy dzięki podziałowi danych według tego atrybutu. Przy wyborze kryterium podziału dążymy do maksymalizacji zysku informacyjnego, co oznacza wybór takiego podziału, który najlepiej redukuje nierównomierność danych. Nasze dane mają różne wartości atrybutu decyzyjnego, a podziały są wynikiem analizy tego atrybutu.)
    
$$
H(X) = -\sum_{i=1}^{n} P(x_i) \log_2(P(x_i))
$$

### Metody "Ensemble"

#### Cel stosowania
Metody "Ensemble" są stosowane w celu rozwiązania problemu wysokiej wariancji występującej w normalnym drzewie decyzyjnym. Tworząc dwa różne drzewa z tego samego zestawu danych, możemy uzyskać dwa predykcje, które znacząco się różnią. Aby poradzić sobie z tym wyzwaniem, opracowano kilka metod, takich jak:

#### Bagging
**Koncepcja**: Jeśli N obserwacji ma wariancję \(\sigma^2\), to wariancja średniej tych obserwacji wynosi \(\frac{\sigma^2}{N}\).

**Metody i charakterystyka**:
- Uprawianie drzew o pełnej długości (bez przycinania).
- Poszczególne drzewa mają wysoką wariancję i niskie odchylenie, a uśrednianie zmniejsza wariancję.
- W regresji bierzemy średnią z przewidywanych wartości.
- W przypadku klasyfikacji, jako ostateczną prognozę przyjmuje się wynik większości, tj. najbardziej przewidywaną klasę.

**Wady**: Bagging tworzy skorelowane drzewa.

#### RandomForest
**Koncepcja**: Używamy podzbioru zmiennych predykcyjnych, aby uzyskać różne podziały w każdym modelu.

**Metody i charakterystyka**:
- Inny zestaw m predyktorów z p. Reguła kciuka dla wartości m:
    - Dla regresji = \(\frac{p}{3}\)
    - Dla klasyfikacji = \(\sqrt{p}\)
- Wiedza biznesowa: Jeśli zmienne są silnie skorelowane, wypróbuj mniejszą wartość m.

#### Boosting
**Koncepcja**: Proces przekształcania słabego ucznia w silnego ucznia, poprzez aplikowanie różnych technik wzmacniających.

**Metody i charakterystyka**:
- Wzmocnienie gradientowe (Gradient Boosting):
    - Pierwsze drzewo - Obliczanie reszt modelu.
    - Drugie drzewo - Drzewo po dostosowaniu do reszt.
    - N-te drzewo: Końcowe drzewo po wielokrotnym dostosowaniu reszt modelu.
- Ada Boost:
    - Pierwsze drzewo - Przypisuje większą wagę błędnie sklasyfikowanym obserwacjom.
    - Drugie drzewo - Ponowne trenowanie drzewa po uwzględnieniu dostosowanych wag.
    - N-te drzewo: Końcowe drzewo po uwzględnieniu dostosowanych wag N-1 razy.
- XG Boost:
    - Podobne do Gradient Boost.
    - XG-boost wykorzystuje bardziej regularną formalizację modelu, aby kontrolować nadmierne dopasowanie, co daje mu lepszą wydajność.
    - W przypadku modelu bardziej odpowiednie może być określenie go mianem regularized gradient boosting.

**Regularyzacja**: Funkcja kosztu, którą staramy się zoptymalizować (MSE w regresji itp.), zawiera również termin kary za liczbę zmiennych. W pewnym sensie chcemy zminimalizować liczbę zmiennych w ostatecznym modelu wraz z MSE lub dokładnością. Pomaga to uniknąć nadmiernego dopasowania. XG-Boost zawiera warunki regularyzacji w funkcji kosztu.

```python
# Define plot_confusion_matrix function
def plot_confusion_matrix(conf_matrix, labels=['Not Fraud', 'Fraud'], title='Confusion Matrix'):
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g',
                xticklabels=labels, 
                yticklabels=labels)
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.title(title)
    plt.show()

# Define EvaluateModel function
def EvaluateModel(y_true, y_pred, y_proba=None, labels=['Not Fraud', 'Fraud']):
    """
    Evaluate the performance of a classification model.

    Parameters:
    - y_true: True labels
    - y_pred: Predicted labels
    - y_proba: Predicted probabilities (for ROC AUC)
    - labels: Class labels
    """
    # Accuracy
    accuracy = accuracy_score(y_true, y_pred)
    print("Accuracy:", accuracy)
    
    # F1 Score
    f1 = f1_score(y_true, y_pred, average='weighted')  # Use 'weighted' for multi-class support
    print("F1 Score (Weighted):", f1)
    
    # Sensitivity (Recall)
    sensitivity = recall_score(y_true, y_pred, average='weighted')
    print("Sensitivity (Recall):", sensitivity)
    
    # Precision
    precision = precision_score(y_true, y_pred, average='weighted')
    print("Precision:", precision)
    
    # Specificity (True Negative Rate)
    conf_matrix = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = conf_matrix.ravel()
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
    print("Specificity (True Negative Rate):", specificity)
    
    # ROC AUC (if probabilities are provided)
    if y_proba is not None:
        roc_auc = roc_auc_score(y_true, y_proba, multi_class='ovr', average='weighted')
        print("ROC AUC Score:", roc_auc)
    
    # Print confusion matrix
    print("Confusion Matrix:")
    print(conf_matrix)
    
    # Plot confusion matrix
    plot_confusion_matrix(conf_matrix, labels)

# Define visualize_distributions function
def visualize_distributions(data, columns, title="Data Distribution"):
    """
    Visualize the distribution of numerical features.
    
    Parameters:
    - data: Pandas DataFrame or NumPy array
    - columns: List of column names for visualization
    - title: Plot title
    """
    plt.figure(figsize=(12, 8))
    for column in columns:
        sns.kdeplot(data[column], label=column)
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

# Example usage:
# y_true = [actual labels]
# y_pred = [predicted labels]
# y_proba = [predicted probabilities]  # Optional
# EvaluateModel(y_true, y_pred, y_proba)

# Visualize distributions
# data = pandas DataFrame or NumPy array
# visualize_distributions(data, columns=['Feature1', 'Feature2'])
```

```python
from scipy.stats import kstest, norm, expon, gamma, lognorm, weibull_min, chi2, uniform, beta

def fit_and_ks_test(column_data):
    # Normal test (estimate mean and std from the data)
    mu, sigma = column_data.mean(), column_data.std()
    ks_normal = kstest(column_data, "norm", args=(mu, sigma))

    # Exponential test (estimate scale from the data)
    loc, scale = 0, 1 / column_data.mean()  # For exponential, scale = 1/lambda
    ks_expon = kstest(column_data, "expon", args=(loc, scale))

    # Gamma test (estimate shape and scale from the data)
    shape, loc, scale = gamma.fit(column_data)
    ks_gamma = kstest(column_data, "gamma", args=(shape, loc, scale))

    # Lognormal test (estimate shape, loc, and scale from the data)
    shape, loc, scale = lognorm.fit(column_data)
    ks_lognorm = kstest(column_data, "lognorm", args=(shape, loc, scale))

    # Weibull test (estimate shape, loc, and scale from the data)
    shape, loc, scale = weibull_min.fit(column_data)
    ks_weibull = kstest(column_data, "weibull_min", args=(shape, loc, scale))

    # Chi-Square test (estimate df from the data)
    df = (column_data.var() / column_data.mean())  # rough estimate for df
    ks_chi2 = kstest(column_data, "chi2", args=(df, loc, scale))

    # Uniform test (estimate loc and scale from the data)
    loc, scale = column_data.min(), column_data.max() - column_data.min()
    ks_uniform = kstest(column_data, "uniform", args=(loc, scale))

    # Beta test (estimate parameters a and b from the data)
    a, b, loc, scale = beta.fit(column_data)
    ks_beta = kstest(column_data, "beta", args=(a, b, loc, scale))

    # Return results as a dictionary
    return {
        "Normal": ks_normal,
        "Exponential": ks_expon,
        "Gamma": ks_gamma,
        "Lognormal": ks_lognorm,
        "Weibull": ks_weibull,
        "Chi-Square": ks_chi2,
        "Uniform": ks_uniform,
        "Beta": ks_beta,
    }
```

*Zanim ocenimy model, ważne jest zauważenie, że nie możemy używać wcześniej utworzonych drzew, ponieważ występuje znaczny brak równowagi klas w danych. Świadczy o tym macierz pomyłek stworzona we wcześniejszych krokach.*

```python
# Load the dataset
df = pd.read_csv("fraud test.csv", header = 0)

# Convert date and time column to datetime format
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'], format = '%Y-%m-%d')

# Extract features from datetime columns
#df['trans_year'] = df['trans_date_trans_time'].dt.year
#df['trans_month'] = df['trans_date_trans_time'].dt.month
#df['trans_day'] = df['trans_date_trans_time'].dt.day
df['trans_hour'] = df['trans_date_trans_time'].dt.hour
#df['trans_minute'] = df['trans_date_trans_time'].dt.minute
#df['trans_second'] = df['trans_date_trans_time'].dt.second
df['dob_year'] = df['dob'].dt.year
#df['dob_month'] = df['dob'].dt.month
#df['dob_day'] = df['dob'].dt.day

# Drop unnecessary columns
df.drop(['Unnamed: 0',
         "first",
         "last",
         "street",
         "city",
         "state",
         "zip",
         "trans_num",
         'trans_date_trans_time',
         "dob",
         "cc_num"
        ],
        axis = 1,
        inplace = True
       )

# Convert categorical variables using one-hot encoding
df = pd.get_dummies(df, columns=['category', 'gender', 'job', "merchant"], drop_first = True)

# X-y split
X = df.loc[:, df.columns!="is_fraud"]
y = df["is_fraud"]

# Undersample the majority class
rus = RandomUnderSampler(random_state = 0)
X_resampled, y_resampled = rus.fit_resample(X, y)

# Test-Train Split on the undersampled data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size = 0.2, random_state = 0)

# Train the decision tree classifier
clftree = tree.DecisionTreeClassifier(max_depth = 3)
clftree.fit(X_train, y_train)

# Predictions
y_test_pred = clftree.predict(X_test)

# Model Evaluation
EvaluateModel(y_test, y_test_pred)
```

Metoda ta polega na tworzeniu wielu niezależnych klasyfikatorów na podstawie losowych próbek ze zbioru treningowego, a następnie łączeniu ich prognoz w celu uzyskania bardziej stabilnych i dokładnych wyników predykcji.*

*Pierwszym krokiem w implementacji tej metody jest instancjonowanie klasyfikatora bazowego, w tym przypadku klasyfikatora drzewa decyzyjnego. Drzewa decyzyjne są popularnym wyborem jako estymatory bazowe ze względu na swoją prostotę i interpretowalność.*

*Następnie, przy użyciu klasy BaggingClassifier z pakietu scikit-learn, tworzony jest klasyfikator baggingowy. W konfiguracji tego klasyfikatora określono, że klasyfikator bazowy będzie stanowił wcześniej zdefiniowany klasyfikator drzewa decyzyjnego. Parametr bootstrap ustawiony na True jest kluczowym aspektem metody baggingu. "Bootstrap" to technika próbkowania z powtórzeniami, która stanowi podstawę tej metody. Główna idea bootstrapu polega na tym, że zamiast korzystać z oryginalnego zbioru danych do trenowania każdego klasyfikatora w ensemble, losujemy z niego próbki z powtórzeniami. Oznacza to, że w każdej próbce mogą znaleźć się duplikaty pewnych instancji danych, a inne mogą zostać pominięte. Dlaczego to jest istotne? Bootstrap daje każdemu klasyfikatorowi w ensemble szansę zobaczenia różnych perspektyw danych treningowych. Dzięki temu każdy model uczy się na nieco innych danych, co prowadzi do zwiększenia różnorodności w ensemble. W kontekście metody baggingu, ustawienie bootstrap=True oznacza, że każdy klasyfikator w ensemble będzie trenowany na losowych próbkach danych pochodzących z oryginalnego zbioru treningowego, z powtórzeniami. To pozwala na wykorzystanie tej samej wielkości próbki co w oryginalnym zbiorze, ale z różnymi zestawami danych dla każdego klasyfikatora. W rezultacie każdy klasyfikator ma szansę "zobaczyć" różne aspekty danych treningowych, co przyczynia się do redukcji wariancji i poprawy ogólnej wydajności modelu ensemble.*

*Liczba estymatorów w zespole została ustalona na 500, a próbkowanie bootstrapowe zostało wykorzystane do losowego wybierania próbek ze zbioru treningowego z powtórzeniami. Ponadto, w celu zwiększenia wydajności obliczeniowej, ustawiono parametr n_jobs na wartość -1, co oznacza wykorzystanie wszystkich dostępnych rdzeni procesora do przetwarzania równoległego. - colsample_bytree: Udział cech używanych do treningu każdego drzewa.
- reg_alpha: Parametr regularyzacji L1, który kontroluje wielkość współczynnika kary dla cech.
